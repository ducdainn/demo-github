```python
list_link_specialization = []
list_name_specialization = []
for link, name in zip(list_link, list_name):
    html = requests.get(link)
    html.text
    bs = BeautifulSoup(html.text, "html.parser")
    get_link = bs.select('div.bt3-container.body-container > div > div > div:nth-child(2) > a')
    for i in get_link:
        href = i.get('href')
        if re.search(r'specializations', href):
            list_link_specialization.append('https://www.coursera.org' + href)
            list_name_specialization.append(name)
```

```python
list_key_course_link = [] # Link key
stars = []
rates = []
levels = []
instructor = []
names = []
videos = []
readings = []
quizzes = []
time = []
# l = ['https://www.coursera.org/specializations/ai-healthcare',
# 'https://www.coursera.org/specializations/algorithms']
# n = ['Stanford University',
#     'Stanford University']
for link, name in zip(list_link_specialization_3, list_name_specialization_3):
    html = requests.get(link)
    html.text
    bs = BeautifulSoup(html.text, "html.parser")
    list_sub_cousre_link = []
    temp_link_course = bs.select('div.css-3glzp7 > div > h3 > a') # Link khóa học nhỏ
    list_star = bs.select('div > div > span:nth-child(5) > span.css-1tdi49m') # Star
    list_rate = bs.select('div > span:nth-child(5) > span.cds-119.css-dmxkm1.cds-121 > span') # Rate
    for sub_link, star, rate in zip(temp_link_course, list_star, list_rate):
        sub_link_ = f"https://www.coursera.org{sub_link.get('href')}"
        html = requests.get(sub_link_)
        html.text
        bs = BeautifulSoup(html.text, "html.parser")
        module_info = bs.select('div.css-15ko5n9 > p > span > span') # info module vd, rd, qiz 
        module_info2 = bs.select('span:nth-child(1) > div > span:nth-child(3) > span')
        module_level = bs.select('div.css-1r9afu3 > section > div.css-lt1dx1 > div > div.cds-119.css-h1jogs.cds-121') # Level
        for i in module_info:
            for j in i: # vd, rd, qiz
                if re.search(r'videos|video', j):
                    videos.append(j)
                elif re.search(r'readings|reading', j):
                    readings.append(j)
                elif re.search(r'quiz|quizzes', j):
                    quizzes.append(j)
                elif re.search(r'hour|hours|minutes|minute', j):
                    time.append(j)
        for i in module_info2:
            for j in i:
                time.append(j)
        level = None
        for i in module_level:
            if re.search(r'level', i.get_text()):
                level = i.get_text()
                break
        checks = bs.select('div.css-1srdv30 > div > p > span > button > span > span')
        check_ins = [True for i in checks if i.text != []]
        temp_list_instructor = []
        if check_ins:
            sub_temp_list_instructor = bs.select('div.css-11krhap div div div div img')
            for i in sub_temp_list_instructor:
                j = i.get('alt')
                if j not in temp_list_instructor:
                    temp_list_instructor.append(j)
        else:
            sub_temp_list_instructor = bs.select('div.css-guxf6x > p > span > a > span')
            for i in sub_temp_list_instructor:
                j = i.get_text()
                if j not in temp_list_instructor:
                    temp_list_instructor.append(j)
        
        max_length = max(len(videos), len(readings), len(quizzes), len(time))

        # Thêm NaN cho các danh sách để có cùng độ dài
        videos.extend([0] * (max_length - len(videos)))
        readings.extend([0] * (max_length - len(readings)))
        quizzes.extend([0] * (max_length - len(quizzes)))
        time.extend([0] * (max_length - len(time)))

        # Tạo danh sách link có cùng độ dài với các danh sách trên
        for i in range(max_length - len(stars)):
            stars.append(star.get_text())
            rates.append(rate.get_text())
            levels.append(level)
            instructor.append(temp_list_instructor)
            names.append(name)
        for i in range(max_length - len(list_key_course_link)):
            list_key_course_link.append(link)
```
